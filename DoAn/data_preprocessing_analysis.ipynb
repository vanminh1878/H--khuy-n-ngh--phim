Ý nghĩa các thuộc tính:

* **Movie_link**: đường dẫn tới trang web có thông tin chi tiết về bộ phim
* **Movie_name**: Tên bộ phim
* **Ratings**: Điểm số đánh giá bộ phim
* **Taglines**: Câu khẩu hiệu hoặc câu châm ngôn quảng cáo được sử dụng để mô tả bộ phim. (3,4)
* **Movie_category**: Thể loại của bộ phim.
* **Film_actor**: Các diễn viên chính hoặc các vai diễn quan trọng trong bộ phim.
* **Release_Date**: Ngày phát hành chính thức của bộ phim.
* **Duration**: Thời lượng của bộ phim, thường được đo bằng phút.
* **Country**: Quốc gia mà bộ phim được sản xuất hoặc liên quan đến. (3,4)
* **Original_Language**: Ngôn ngữ gốc được sử dụng trong bộ phim.
* **Companies**: Các công ty hoặc hãng sản xuất, phân phối, hoặc hỗ trợ sản xuất bộ phim. (3)
* **Description**: Mô tả ngắn gọn về nội dung của bộ phim.
* **Budget**: Ngân sách sản xuất của bộ phim, tức là số tiền đã chi tiêu để sản xuất. (4)
* **Box Office**: Doanh thu thu được từ việc chiếu phim, thường được đo trong khoản tiền.
* **Certificate (MPAA)**: Chứng nhận phim của Hiệp hội Phim Mỹ (MPAA), đánh giá mức độ phù hợp với khán giả ở mọi độ tuổi. (3)
* **Directors**: Đạo diễn hoặc các nhà làm phim chịu trách nhiệm chỉ đạo quá trình sản xuất và sáng tác của bộ phim.
* **Writing_Credits**: Tác giả hoặc các nhà biên kịch đóng góp vào kịch bản hoặc cốt truyện của bộ phim.
* **poster_path**: Đường dẫn tới poster bộ phim.

Có 5 cột không tồn tại ở cả 4 datasets là: Taglines, Country, Companies, Budget, Certificate (MPAA). Mục đích sử dụng của bộ dữ liệu là để đào tào mô hình khuyến nghị phim, do đó, chúng tôi tự đánh giá theo kinh nghiệm và thực hiện xóa 3 cột là Taglines, Companies, Budget.
